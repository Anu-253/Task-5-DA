```python
import pandas as pd  #importing libraries
import seaborn as sns
import matplotlib.pyplot as plt

# To make plots look nicer
sns.set(style="whitegrid")
```

```python
# Load Titanic dataset
df = pd.read_csv("train.csv")   # change path if needed
```

```python

# If Titanic dataset is in the same folder as your notebook
df = pd.read_csv("train.csv")

# Quick check
df.head()
```

```python
print("Shape of dataset:", df.shape) #finding shape of dataset
print("\nInfo about dataset:")
print(df.info())
print("\nSummary statistics:")
print(df.describe(include="all"))

# Survival counts
print("\nSurvival distribution:")
print(df['Survived'].value_counts())
```

```python
#Age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#survival by gender
plt.figure(figsize=(6,5))
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
#survival for passesnger class
plt.figure(figsize=(6,5))
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
```

```python
#Observations from Titanic EDA
1. Age Distribution

Most passengers were between 20 and 40 years old.

There were also a noticeable number of children and teenagers, but fewer elderly passengers.

The dataset has some missing values in Age, which could affect analysis if not handled.

2. Survival by Gender

Females had a much higher survival rate compared to males.

This aligns with the "women and children first" rescue policy followed during the disaster.

Most of the male passengers did not survive.

3. Survival by Passenger Class

First-class passengers survived more often compared to second and third class.

Third-class passengers had the lowest survival rate, showing that socio-economic status influenced survival chances.

Wealthier passengers (1st class) had better access to lifeboats.
```
